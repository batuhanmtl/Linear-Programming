maximize $Z = 3X_1+X_2$

subject to

- $X_2 \leqslant 5$
- $X_1 + X_2  \leqslant 10$
- $X_1 - X_2  \leqslant 2$
- $X_1 , X_2 \geqslant 0$


With the linprog method, the minimization is calculated, but we are doing the maximization, so we must multiply the coefficients in the objective function by -1.
We will solve the problem with the simplex method.

```python
from scipy.optimize import linprog
import numpy as np

A = np.array([[0,1],[1,1],[1,-1]])
b = np.array([5,10,2])
c = np.array([-3,-1])

result = linprog(c=c,A_ub=A,b_ub=b,method='simplex',bounds=(0,None))

result.fun*(-1)
```

Since we multiply by -1 first, we need to multiply the resulting value by -1 again for the correct result.
So the maximum value is 22.0

```python
result.x
```

X1 value was found as 6.0 and X2 value as 4.0.

Let's graph the constraint functions to see the appropriate solution area.

```python
import matplotlib.pyplot as plt

X = np.linspace(0,10,10) #trial values for the chart.
y1 = 5-0*X
y2 = 10-X
y3=X-2

plt.plot(X,y1,label='$y=5-0*x_1$')
plt.plot(X,y2,label='$y=10-x_1$')
plt.plot(X,y3,label='$y=x_1-2$')
plt.scatter(result.x[0],result.x[1],color='y')
plt.legend()
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.style.use('ggplot')
plt.show()
```

Now let's plot the graph including the objective function.
- objective function : $y=22-3x_1$


```python
y4 = 22-3*X # Z function
plt.plot(X,y1,label='$y=5-0*x_1$')
plt.plot(X,y2,label='$y=10-x_1$')
plt.plot(X,y3,label='$y=x_1-2$')
plt.plot(X,y4,label='$Z=22-3x_1$',linestyle='--')
plt.scatter(result.x[0],result.x[1],color='orange')
plt.legend()
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.xlim(0)
plt.ylim(0,10)
plt.style.use('ggplot')
plt.show()
```
